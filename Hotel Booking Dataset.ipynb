```python
import pandas as pd

# Load data from Excel
main_df = pd.read_excel("C:/Users/Admin/Downloads/Nikhil Sharma files/Hotel Booking Dataset/Hotel Booking Analytics_Dataset.xlsx", sheet_name="Hotel_Booking")
room_ref = pd.read_excel("C:/Users/Admin/Downloads/Nikhil Sharma files/Hotel Booking Dataset/Hotel Booking Analytics_Dataset.xlsx", sheet_name="Room_Type")
meal_ref = pd.read_excel("C:/Users/Admin/Downloads/Nikhil Sharma files/Hotel Booking Dataset/Hotel Booking Analytics_Dataset.xlsx", sheet_name="Meal_Type")
Country_ref = pd.read_excel("C:/Users/Admin/Downloads/Nikhil Sharma files/Hotel Booking Dataset/Hotel Booking Analytics_Dataset.xlsx", sheet_name="Country")

```

```python
# Merge room type reference and replace reserved room code with full name
main_df = main_df.merge(room_ref, how='left', left_on='reserved_room_type_code', right_on='room_code')
main_df.drop(columns=['reserved_room_type_code', 'room_code'], inplace=True)
main_df.rename(columns={'room_type_name': 'reserved_room_type'}, inplace=True)

# Merge meal type reference and replace meal code with full name
main_df = main_df.merge(meal_ref, how='left', on='meal_code')
main_df.drop(columns=['meal_code'], inplace=True)
main_df.rename(columns={'meal_type_name': 'meal_type'}, inplace=True)

# Merge country reference and replace country code with full name
main_df = main_df.merge(Country_ref, how='left', on='country_code')
main_df.drop(columns=['country_code'], inplace=True)
main_df.rename(columns={'country_name': 'country'}, inplace=True)
```
